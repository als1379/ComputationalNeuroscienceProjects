```python
from STDP_Learning import spike_time_dependant_learning
from NeuronGroups import *
import math
```

```python
func1 = lambda x: 200 * (math.sqrt(math.fabs(math.sin(x))))
func2 = lambda x: 250 * math.sin(x)

neuron1 = LIF(input_current=func1)
neuron2 = LIF(input_current=func1)
neuron3 = LIF(input_current=func2)

connections = [[1, 2], [0, 2], [0, 1]]
neurons = [neuron1, neuron2, neuron3]

weight_table = []

for i in range(3):
    temp = []
    for j in range(3):
        temp.append(random.randint(8, 14))
    weight_table.append(temp)

for i in range(3):
    weight_table[i][i] = 0
print(weight_table)

ng = NeuronsGroup(neurons=neurons, connections=connections, weight_table=weight_table)
STDP = spike_time_dependant_learning(neurons_group=ng, amplify_val_pos=10, amplify_val_neg=-10, delta_tau_pos=0.2, delta_tau_neg=0.2)
STDP.train()
```
