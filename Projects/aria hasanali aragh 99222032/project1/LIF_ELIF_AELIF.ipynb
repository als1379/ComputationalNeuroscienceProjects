```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D
import math
```

```
class LIF():


    def __init__(self, I, Time , time_delta,treshold, u_rest, u_reset, R, tau):
        self.I = I
        self.Time = Time
        self.time_delta = time_delta
        self.u_history = []
        self.fire_time_history = []
        self.u = u_rest
        self.treshold = treshold
        self.u_rest = u_rest
        self.u_reset = u_reset
        self.R = R
        self.tau = tau


    def start(self):
        current_time = 0
        while True:    
            du = (-1 * (self.u - self.u_rest) + self.R * self.I(current_time)/ 1000) / self.tau
            self.u += du * self.time_delta
            if self.u >= self.treshold:
                self.fire_time_history.append(current_time)
                self.u = self.u_reset
            self.u_history.append(self.u)
            current_time += self.time_delta
            if current_time > self.Time:
                break
        return self.u_history


    def show_plot(self):
        plt.plot(range(len(self.u_history)), self.u_history)
        plt.show()

```

```
# constant input

def I(x):
  return 2000
```

```
lif = LIF(I, treshold=-55, u_rest=-70, u_reset=-65, R=14, tau=28, Time=200, time_delta=0.001)
lif.start()
lif.show_plot()
```

```
# sin input 
def I(x):
  return 4000 * (math.sin(x) + 0.9)
```

```
lif = LIF(I, treshold=-55, u_rest=-70, u_reset=-65, R=14, tau=28, Time=200, time_delta=0.001)
lif.start()
lif.show_plot()
```

```
class ELIF():


    def __init__(self, I, Time, time_delta ,treshold, u_rest, u_reset, R, tau, delta_T, theta):
        self.I = I
        self.Time = Time
        self.time_delta = time_delta
        self.u_history = []
        self.fire_time_history = []
        self.u = u_rest
        self.treshold = treshold
        self.u_rest = u_rest
        self.u_reset = u_reset
        self.R = R
        self.tau = tau
        self.delta_T = delta_T
        self.theta = theta



    def start(self):
        current_time = 0
        while True:    
            du = (-1 * (self.u - self.u_rest) + self.delta_T * math.exp((self.u - self.theta)/self.delta_T) + self.R * self.I(current_time)/ 1000) / self.tau
            self.u += du * self.time_delta
            if self.u >= self.treshold:
                self.fire_time_history.append(current_time)
                self.u = self.u_reset
            self.u_history.append(self.u)
            current_time += self.time_delta
            if current_time > self.Time:
                break
        return self.u_history


    def show_plot(self):
        plt.plot(range(len(self.u_history)), self.u_history)
        plt.show()
```

Adaptive ELIF

```
class AELIF():


    def __init__(self, I, Time, time_delta,treshold, u_rest, u_reset, R, tau, delta_T, theta, a, b, tau_w):
        self.I = I
        self.Time = Time
        self.time_delta = time_delta
        self.u_history = []
        self.w_history = []
        self.fire_time_history = []
        self.u = u_rest
        self.treshold = treshold
        self.u_rest = u_rest
        self.u_reset = u_reset
        self.R = R
        self.tau = tau
        self.delta_T = delta_T
        self.theta = theta
        self.a = a
        self.b = b
        self.tau_w = tau_w
        self.w = 0



    def start(self):
        current_time = 0
        while True:    
            dw = ((self.a * (self.u - self.u_rest)) - self.w) / self.tau_w + self.b * len(self.fire_time_history)
            self.w += dw * self.time_delta
            du = (-1 * (self.u - self.u_rest) + self.delta_T * math.exp((self.u - self.theta)/self.delta_T) - self.R * self.w / 1000 + self.R * self.I(current_time)/ 1000) / self.tau
            self.u += du * self.time_delta
            if self.u >= self.treshold:
                self.u = self.u_reset
                self.fire_time_history.append(current_time)
            self.u_history.append(self.u)
            self.w_history.append(self.w)
            current_time += self.time_delta
            if current_time > self.Time:
                break
        return self.u_history


    def show_plot_ut(self):
        plt.plot(range(len(self.u_history)), self.u_history)
        plt.show()

    def show_plot_wt(self):
        plt.plot(range(len(self.w_history)), self.w_history)
        plt.show()
```

```
# constant input

def I(x):
  return 2000
```

```
aelif = AELIF(I=I, treshold=-50, u_rest=-70, u_reset=-65, R=10, tau=8, Time=100, time_delta=0.1, 
    delta_T=1, theta=-60,
     a=0.01, b=100, tau_w=1)
aelif.start()
aelif.show_plot_ut()
```
